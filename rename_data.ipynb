```python
import os
import shutil

folder_path = 'C:\\Users\\A\\Desktop\\224_Data\\R224_Xmin0_Xmax2_Ymin0_Ymax2_e-10\\R224_Xmin0_Xmax2_Ymin0_Ymax2_e10'
start_number = 13161
```

```python

def sort_by_number(filename):
    # 파일 이름에서 숫자 부분 추출
    number_part = ''.join(filter(str.isdigit, filename))
    return int(number_part) if number_part.isdigit() else float('inf')

# 파일을 숫자 순서대로 정렬
sorted_filenames = sorted(os.listdir(folder_path), key=sort_by_number)

for filename in sorted_filenames:
    print(filename)
```

```python
for filename in sorted_filenames:
    old_file = os.path.join(folder_path, filename)

    if os.path.isfile(old_file):
        _, extension = os.path.splitext(filename)
        new_filename = f"{start_number}{extension}"
        new_file = os.path.join(folder_path, new_filename)

        shutil.move(old_file, new_file)
        start_number += 1
```

```python
import os
import numpy as np

data_folder =folder_path

def load_and_combine_data(data_folder):
    file_paths = [os.path.join(data_folder, f) for f in os.listdir(data_folder) if f.endswith('9.txt')]
    all_data = []
    expected_shape = None  # 초기에는 기대하는 형태를 설정하지 않음

    for file_path in file_paths:
        data = np.loadtxt(file_path)
        
        if expected_shape is None:
            expected_shape = data.shape  # 첫 번째 파일의 형태를 기대하는 형태로 설정
        elif data.shape != expected_shape:
            print(f"형태가 다른 파일 발견: {file_path}, 형태: {data.shape}, 기대하는 형태: {expected_shape}")
            continue  # 형태가 다른 파일은 건너뛰고 다음 파일로 진행

        all_data.append(data)

    combined_data = np.concatenate(all_data, axis=0)
    return combined_data

combined_data = load_and_combine_data(data_folder)
```
